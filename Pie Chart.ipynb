```python
import matplotlib.pyplot as plt
```

```python
cla=['Python','R','Machine Learning','Artificial intelligence','Data Science']
cla1st=[46,78,30,43,28]
te={"fontsize":14}
plt.pie(cla1st,labels=cla,explode=[0,0,0.2,0,0.1],colors=['r','b','g','y','c'],autopct="%0.1f%%",textprops=te,pctdistance=0.4,radius=1.8,startangle=87,rotatelabels=True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
wed={"linewidth":4,"width":3,"edgecolor":"k"}
te={"fontsize":14}
plt.pie(cla1st,labels=cla,explode=[0,0,0.2,0,0.1],colors=['r','b','g','y','c'],autopct="%0.1f%%",textprops=te,wedgeprops=wed,pctdistance=0.4,radius=1.8,startangle=87,rotatelabels=True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
wed={"linewidth":4,"width":3,"edgecolor":"k"}
te={"fontsize":14}
plt.pie(cla1st,labels=cla,explode=[0,0,0.2,0,0.1],colors=['r','b','g','y','c'],autopct="%0.1f%%",textprops=te,wedgeprops=wed,pctdistance=0.4,radius=1.8,startangle=87,rotatelabels=True)
plt.legend(loc=1)
plt.show()
```
