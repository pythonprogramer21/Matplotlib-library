```python
from matplotlib import style
day=['monday','tuesday','wenesday','thurstday','friday','saturday','sonday']
tem=[35.6,29,40,11,32,43,10]
style.use('ggplot')
plt.plot(day,tem,color='c',marker='o',linestyle='--')
plt.title('Maharastra Temperature')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(['Temp line'],loc=4)
plt.show()
```

```python
from matplotlib import style
day=['monday','tuesday','wenesday','thurstday','friday','saturday','sonday']
tem=[35.6,29,40,11,32,43,10]
style.use('ggplot')
plt.plot(day,tem,color='m',marker='o',linestyle='--',linewidth=5)
plt.title('Maharastra Temperature')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(['Temp line'],loc=4)
plt.grid(color='r',linestyle='-',linewidth=3)
plt.show()
```

```python

day=['monday','tuesday','wenesday','thurstday','friday','saturday','sonday']
tem=[35.6,29,40,11,32,43,10]
delhi=[50,2,43,32,5,29,36]

plt.plot(day,tem,color='m',marker='o',linestyle='--',linewidth=5,label='Maharastra')
plt.plot(day,delhi,color='y',marker='o',linestyle=':',linewidth=5,label='Delhi')
plt.title('Maharastra And deli Temperature')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(loc=4)

plt.show()
```
