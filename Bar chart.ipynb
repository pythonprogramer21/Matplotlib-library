```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
cla=['Python','R','AI','ML','DS']
cla1st=[30,10,20,25,10]
cla2st=[40,5,20,20,10]
cla3st=[35,5,30,15,15]
```

```python
plt.bar(cla,cla1st)
```

```python
plt.barh(cla,cla1st)
```

```python
plt.bar(cla,cla1st,width=.3)
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r')
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b')
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2)
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2,alpha=.6)
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--')
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=True)
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=False)
```

```python
plt.bar(cla,cla1st,width=.3,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=True,label='Class 1')
plt.legend(loc=1)
```

```python
style.use('ggplot')
plt.bar(cla,cla1st,width=.6,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=True,label='Class 1')
plt.legend(loc=1)
```

```python
plt.figure(figsize=(13,6))
plt.bar(cla,cla1st,width=.6,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=True,label='Class 1')
plt.legend(loc=1)
```

```python
plt.figure(figsize=(13,6))
plt.bar(cla,cla1st,width=.6,align='edge',color='r',edgecolor='b',linewidth=2,linestyle='--',visible=True,label='Class 1')
plt.legend(loc=1)
plt.title('This is student info')
plt.xlabel('Courses')
plt.ylabel('Student Size')
plt.show()
```

```python
cla_in=np.arange(len(cla))
width=0.2
plt.bar(cla_in,cla1st,width,color='b',label='Class 1')
plt.bar(cla_in+width,cla2st,width,color='g',label='Class ')
plt.bar(cla_in+width+width,cla3st,width,color='y',label='Class 1')
plt.xticks(cla_in+width,cla)
plt.title('This is student info')
plt.xlabel('Courses')
plt.ylabel('Student Size')
plt.legend(loc=1)
plt.show()
```

```python
cla_in=np.arange(len(cla))
width=0.3
plt.bar(cla_in,cla1st,width,color='b',label='Class 1')
plt.bar(cla_in+width,cla2st,width,color='g',label='Class 2')
plt.bar(cla_in+width+width,cla3st,width,color='y',label='Class 3')
plt.xticks(cla_in+width,cla)
plt.title('This is student info')
plt.xlabel('Courses')
plt.ylabel('Student Size')
plt.legend(loc=1)
plt.show()
```

```python
cla_in=np.arange(len(cla))
width=0.3
plt.bar(cla_in,cla1st,width,color='b',label='Class 1')
plt.bar(cla_in+width,cla2st,width,color='g',label='Class 2')
plt.bar(cla_in+width+width,cla3st,width,color='y',label='Class 3')
plt.xticks(cla_in+width,cla,rotation=13)
plt.title('This is student info')
plt.xlabel('Courses')
plt.ylabel('Student Size')
plt.legend(loc=1)
plt.show()
```
