```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
mlst=np.random.randint(18,45,(20))
plst=np.random.randint(15,40,(30))
bins=[15,20,30,45]
plt.hist(plst,bins,rwidth=0.7,color='y',label='PL',orientation='horizontal')
plt.hist(mlst,bins,rwidth=0.7,color='m',label='ML')
plt.title('pL student')
plt.legend(loc=1)
plt.show()
```

```python
mlst=np.random.randint(18,45,(20))
plst=np.random.randint(15,40,(30))
bins=[15,20,30,45]
plt.hist([plst,mlst],bins,rwidth=0.7,color=['y','m'],label=['pl stu','ml stu'])

plt.title('pL student')
plt.legend()
plt.show()
```

```python
mlst=np.random.randint(18,45,(20))
plst=np.random.randint(15,40,(30))
bins=[15,20,30,45]
plt.figure(figsize=(12,9))
plt.hist([plst,mlst],bins,rwidth=0.7,color=['y','m'],label=['pl stu','ml stu'])

plt.title('pL student')
plt.legend()
plt.show()
```

```python
from  matplotlib import style
mlst=np.random.randint(18,45,(20))
plst=np.random.randint(15,40,(30))
bins=[15,20,30,45]
style.use('ggplot')
plt.figure(figsize=(12,9))
plt.hist([plst,mlst],bins,rwidth=0.7,color=['y','m'],label=['pl stu','ml stu'])

plt.title('pL student')
plt.legend()
plt.show()
```
