```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
k=pd.read_csv('F:\\suryakant notes\\pra1.csv')
k
```

```python
x=k['Profit']
y=k['Rate']
plt.scatter(x,y)
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r")
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",marker='+')
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",marker='*',s=130)
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",marker='*',s=130,alpha=.3)
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",marker='*',s=130,linewidth=10)
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",verts="<",s=130,linewidth=6,edgecolor='b')
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y,c="r",marker="<",s=130,linewidth=6,edgecolor='b')
plt.scatter(x,k['Rate1'],c='b',marker="*",s=130,linewidth=6,edgecolor='b')
plt.title('Copany data')
plt.xlabel('Profit')
plt.ylabel('Rate')
plt.show()
```
